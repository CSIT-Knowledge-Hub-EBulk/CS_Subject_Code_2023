## Algorithm:  Factoradic

```python
import time
```

# Algorithm

```python
%%time
fac = lambda i, *j: i and fac(*divmod(i, len(j) + 1), *j) or j or (i,)
```

```python
%%time
dec = lambda i, k=0, *j: j and dec(i * len(j) + i + k, *j) or i
```

# Run

```python
%%time
for i in range(0, 11):
    f = fac(i ** 3)
    d = dec(*f)
    print(d, '<->', ' '.join(map(str, f)))
```
