## Algorithm:  Shuffle

```python
import time
```

```python
import numpy as np
```

# Algorithm

```python
%%time
def shuffle(data):
    n = len(data)
    for i in range(n):
        k = np.random.randint(i, n)
        data[i], data[k] = data[k], data[i]

    return data
```

# Run

## Shuffle

```python
%%time
data = list(range(10))
```

```python
%%time
shuffle(data)
```

```python
%%time
shuffle(data)
```
