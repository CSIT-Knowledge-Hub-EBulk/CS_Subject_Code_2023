## Algorithm: Hanoi Tower

```python
import time
```

# Algorithm

```python
%%time
def hanoi(height, left='left', right='right', middle='middle'):
    if height:
        hanoi(height - 1, left, middle, right)
        print(left, '=>', right)
        hanoi(height - 1, middle, right, left)
```

# Run

```python
%%time
hanoi(1)
```

```python
%%time
hanoi(2)
```

```python
%%time
hanoi(3)
```
