```python
%%time
def modinv(x, y):
    r, s = 1, 0
    while y:
        r, s = s, r - x // y * s
        x, y = y, x % y
    return r
```

```python
%%time
def rsa_generate_keys():
    p, q = secure_prime_generator()
    n = p * q
    t = n - p - q + 1
    e = 65537
    d = modinv(e, t) % t
    assert (d * e) % t == 1
    return (e, n), (d, n)
```

```python
%%time
def rsa(plaintext, public_key):
    return pow(plaintext, *public_key)
```

```python
%%time
public_key, secret_key = rsa_generate_keys()
```
