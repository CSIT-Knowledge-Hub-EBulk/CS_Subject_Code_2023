## Algorithm:  Gram Schmidt

```python
import time
```

```python
import numpy as np
```

# Algorithm

```python
%%time
def gram_schmidt(X):
    O = np.zeros(X.shape)

    for i in range(X.shape[1]):
        # orthogonalization
        vector = X[:, i]
        space = O[:, :i]
        projection = vector @ space
        vector = vector - np.sum(projection * space, axis=1)

        # normalization
        norm = np.sqrt(vector @ vector)
        vector /= abs(norm) < 1e-8 and 1 or norm

        O[:, i] = vector

    return O
```

# Run

```python
%%time
# 6 column vectors in 4D
vectors = np.array([
    [1, 1, 2, 0, 1, 1],
    [0, 0, 0, 1, 2, 1],
    [1, 2, 3, 1, 3, 2],
    [1, 0, 1, 0, 1, 1]
], dtype=float)
```

```python
%%time
# check orthogonality
vectors.T @ vectors
```

```python
%%time
orthonormal = gram_schmidt(vectors)
orthonormal.round(5)
```

```python
%%time
# check orthogonality
(orthonormal.T @ orthonormal).round(5)
```

## QR decomposition

```python
%%time
matrix = np.array([
    [1, 1, -1],
    [1, 2, 1],
    [1, 3, 0]
], dtype=float)
```

```python
%%time
Q = gram_schmidt(matrix)
Q.round(5)
```

```python
%%time
R = Q.T @ matrix
R.round(5)
```

```python
%%time
 (Q @ R).round(5)
```
