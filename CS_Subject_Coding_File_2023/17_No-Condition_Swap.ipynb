## Algorithm: No-Condition Swap

```python
import time
```

# Algorithm

this is C code
void swap(int *x, int *y) {
    int u = *x, v = *y;
    int s = (u - v) >> (sizeof(int) * 8 - 1);
    *x = v * (1 + s) - u * s;
    *y = u * (1 + s) - v * s;
}
this is Python equivalent

```python
%%time
def swap(x, y):
    s = x < y
    return x * s + y * (1 - s), y * s + x * (1 - s)
```

# Run

```python
%%time
swap(3, 15), swap(15, 3)
```

```python
%%time
swap(-13, 5), swap(5, -13)
```
