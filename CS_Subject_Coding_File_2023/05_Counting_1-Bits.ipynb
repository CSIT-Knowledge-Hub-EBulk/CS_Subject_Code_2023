## Algorithm: Counting 1-Bits

```python
import time
```

# Algorithm

```python
%%time
def count_of_1bits(value):
    n = 0
    while value:
        value &= value - 1
        n += 1
    return n
```

# Run

```python
%%time
count_of_1bits(0)
```

```python
%%time
count_of_1bits(1)
```

```python
%%time
count_of_1bits(0b11001100)
```
